```python
dfnull = df.isna()
dfmissing = dfnull.sum()
percentagemissing = (dfmissing / len(df))*100
print(percentagemissing)
```

```python
#Dropping columns
df.drop(columns = ['ADDRESSLINE2','STATE','TERRITORY','PHONE','ADDRESSLINE1','POSTALCODE','CONTACTFIRSTNAME','CONTACTLASTNAME'], inplace = True)
```

```python
dfnull = df.isna()
dfmissing = dfnull.sum()
percentagemissing = (dfmissing / len(df))*100
print(percentagemissing)
```

```python
import matplotlib.pyplot as plt 
df1=df.select_dtypes(exclude=['object'])
for column in df1:
        plt.figure(figsize=(17,1))
        sns.boxplot(data=df1, x=column)
```

```python
#Datatype detection
consistency_results ={}

for column in df.columns:
    unique_types = df[column].map(type).unique()
    
    if len(unique_types) == 1:
        consistency_results[column] = True
    else:
        consistency_results[column] = False
        
print(consistency_results)
```

```python
df.head(20)
```

```python
##Total Sales by Region

df2 = df.groupby('COUNTRY')['SALES'].sum().reset_index()
df2 = df2.sort_values(by = 'SALES', ascending = False)
df2
```

```python
plt.figure(figsize = (10,7))
plt.bar(df2['COUNTRY'], df2['SALES'], color = 'hotpink',width = 0.3)
plt.title("Total Sales Across Different Countries", size = 20)
plt.xlabel("Country", size = 20)
plt.ylabel("Sales", size = 20)
plt.xticks(rotation = 45)
plt.show()
```

```python
##Total Sales by product Line
```

```python
df3 = df.groupby('PRODUCTLINE')['SALES'].sum().reset_index()
df3.sort_values(by = 'SALES', ascending = False)
```

```python
plt.figure(figsize = (10,7))
plt.bar(df3['PRODUCTLINE'], df3['SALES'],width = 0.3)
plt.title("Total Sales by Product Line", size = 20)
plt.xlabel("Product Line", size = 20)
plt.ylabel("Sales", size = 20)
plt.xticks(rotation = 45)
plt.show()
```

```python
productline_counts = df['PRODUCTLINE'].value_counts()
colors = ['Voilet','Indigo','Green','Blue','Yellow', 'Orange']
plt.figure(figsize=(6, 10))
plt.pie(productline_counts, labels=productline_counts.index, autopct='%1.2f%%', colors=sns.color_palette('Set2'))
plt.title('Count of Orders by Product Line', size = 20)
plt.show()
```

```python
df4 = df.groupby('YEAR_ID')['SALES'].sum().reset_index()
df4['YEAR_ID'] = df4['YEAR_ID'].astype(int)

plt.figure(figsize = (10,7))
plt.plot(df4['YEAR_ID'], df4['SALES'], color = 'red', linewidth = 5, marker = 'o', markersize = 10)

plt.title("Sales Variation", size = 20)
plt.xlabel("Year", size = 20)
plt.ylabel("Sales", size = 20)
plt.xticks(ticks=df4['YEAR_ID'], labels=df4['YEAR_ID'].astype(str))
plt.show()
```

```python
df5 = df.groupby('CITY')['SALES'].sum().reset_index()
df5 = df5.sort_values(by = 'SALES', ascending = False)
df6 = df5.head(10)

plt.figure(figsize=(15, 7))
colors = plt.cm.Blues(df5['SALES'] / df5['SALES'].max())
bars = plt.bar(df5['CITY'], df5['SALES'], color=colors)

plt.title("Top 10 Cities By Sales", size=20)
plt.xlabel("City", size=15)
plt.ylabel("Sales", size=15)
current_values = plt.gca().get_yticks()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sizes = df['DEALSIZE'].value_counts()
colors = ['blue', 'green', 'orange']
plt.figure(figsize=(10, 7))
plt.pie(sizes, labels=sizes.index, autopct='%1.1f%%',startangle = 140)
plt.axis('equal')
plt.title('Distribution of Small, Medium, and Large Dealsizes')
plt.show()
```

```python
grouped_data = df.groupby(['COUNTRY', 'DEALSIZE'])['SALES'].sum().unstack()
colors = ['lightblue', 'lightgreen', 'orange'] 
grouped_data = grouped_data.loc[grouped_data.sum(axis=1).sort_values(ascending=False).index]
grouped_data.plot(kind='bar', stacked=True, color=colors, figsize=(10, 6))

plt.title('Sales by Country and Dealsize')
plt.xticks(rotation = 45)
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.show()
```
